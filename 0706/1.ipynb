```python
import math

# Circle 클래스 정의
class Circle:
    def __init__(self, radius = 0):
        self.radius = radius

    def getArea(self):
        return math.pi * (self.radius ** 2)

    def getPerimeter(self):
        return 2 * math.pi * self.radius
    
# Circle 객체 생성
c1 = Circle(10)
print("원의 면적c1:", c1.getArea())
print("원의 둘레c1:", c1.getPerimeter())
c2 = Circle(5)
print("원의 면적c2:", c2.getArea())
print("원의 둘레c2:", c2.getPerimeter())
```

```python
class Car:
    def __init__(self, speed, color, model):
        self.speed = speed
        self.color = color
        self.model = model

    def get_speed(self, speed):
        self.speed = speed
    
    def get_speed(self):
        return self.speed
    
    def __str__(self):
        return f"Car(speed={self.speed}, color={self.color}, model={self.model})"
    
car1 = Car(100, "red", "sedan")
car2 = Car(200, "blue", "SUV")
print(car1)
print(car2)

```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def get_area(self):
        return self.width * self.height
 
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"
    
    def __add__(self, other):
        new_width = self.width + other.width
        new_height = self.height + other.height
        return Rectangle(new_width, new_height)
    
rect1 = Rectangle(10, 20)
rect2 = Rectangle(30, 40)
rect3 = rect1 + rect2
print(rect3)
    
```

```python
from random import randint

class Dice:
    
    def __init__(self, x, y):
        self._x = x
        self._y = y
        self._size = 30
        self.value = 1

    def read_dice(self):
        return self.value
    
    def print_dice(self):
        print(f"Dice value: {self.value}")
        
    def roll_dice(self):
        self.value = randint(1, 6)

d = Dice(100, 100)
d.roll_dice()
d.print_dice()
```

```python
class Car:
    def __init__(self, make, model, color, speed):
        self.make = make
        self.model = model
        self.color = color
        self.speed = speed

    def set_make(self, make):
        self.make = make
    
    def get_make(self):
        return self.make
    
    def __str__(self):
        return f"Car(make={self.make}, color={self.color}, model={self.model})"
    
    
class ElectricCar(Car):
    def __init__(self, make, model, color, price, battery_capacity):
        super().__init__(make, model, color, price)
        self.battery_capacity = battery_capacity

    def set_battery_capacity(self, battery_capacity):
        self.battery_capacity = battery_capacity
    
    def get_battery_capacity(self):
        return self.battery_capacity
    
    def __str__(self):
        return f"ElectricCar(make={self.make}, color={self.color}, model={self.model}, battery_capacity={self.battery_capacity})"
    
car1 = Car("Toyota", "Camry", "red", 100)
car2 = Car("Honda", "Civic", "blue", 200)
electric_car = ElectricCar("Tesla", "Model 3", "black", 50000, 100)

print(car1)
print(car2)
print(electric_car)
myCar = ElectricCar("Nissan", "Leaf", "white", 30000, 40)
print(myCar)
```

```python
class Shape:
    def __init__(self, name):
        self.name = name
    def area(self):
        raise NotImplementedError("Subclasses must implement this method")
    
class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)
    
class Rectangle(Shape):
    def __init__(self, width, height):
        super().__init__("Rectangle")
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
    
shapeList = [Circle(5), Rectangle(4, 6), Circle(3), Rectangle(2, 8)]
for shape in shapeList:
    print(f"{shape.name} area: {shape.area()}")
```

```python
import tkinter as tk

def button_click(value):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, current + value)

def clear():
    entry.delete(0, tk.END)

def calculate():
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")


window = tk.Tk()
window.title("계산기")
window.geometry("400x500")

entry = tk.Entry(window, font=("Arial", 24), justify="right")
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

btn7 = tk.Button(window, text="7", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("7"))
btn7.grid(row=1, column=0)

btn8 = tk.Button(window, text="8", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("8"))
btn8.grid(row=1, column=1)

btn9 = tk.Button(window, text="9", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("9"))
btn9.grid(row=1, column=2)

btn_div = tk.Button(window, text="/", font=("Arial", 20), width=5, height=2,
                    command=lambda: button_click("/"))
btn_div.grid(row=1, column=3)


btn4 = tk.Button(window, text="4", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("4"))
btn4.grid(row=2, column=0)

btn5 = tk.Button(window, text="5", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("5"))
btn5.grid(row=2, column=1)

btn6 = tk.Button(window, text="6", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("6"))
btn6.grid(row=2, column=2)

btn_mul = tk.Button(window, text="*", font=("Arial", 20), width=5, height=2,
                    command=lambda: button_click("*"))
btn_mul.grid(row=2, column=3)


btn1 = tk.Button(window, text="1", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("1"))
btn1.grid(row=3, column=0)

btn2 = tk.Button(window, text="2", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("2"))
btn2.grid(row=3, column=1)

btn3 = tk.Button(window, text="3", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("3"))
btn3.grid(row=3, column=2)

btn_minus = tk.Button(window, text="-", font=("Arial", 20), width=5, height=2,
                      command=lambda: button_click("-"))
btn_minus.grid(row=3, column=3)


btn0 = tk.Button(window, text="0", font=("Arial", 20), width=5, height=2,
                 command=lambda: button_click("0"))
btn0.grid(row=4, column=0)

btn_dot = tk.Button(window, text=".", font=("Arial", 20), width=5, height=2,
                    command=lambda: button_click("."))
btn_dot.grid(row=4, column=1)

btn_clear = tk.Button(window, text="C", font=("Arial", 20), width=5, height=2,
                      command=clear)
btn_clear.grid(row=4, column=2)

btn_plus = tk.Button(window, text="+", font=("Arial", 20), width=5, height=2,
                     command=lambda: button_click("+"))
btn_plus.grid(row=4, column=3)


btn_equal = tk.Button(window, text="=", font=("Arial", 20), width=22, height=2,
                      command=calculate)
btn_equal.grid(row=5, column=0, columnspan=4)

window.mainloop()
```
